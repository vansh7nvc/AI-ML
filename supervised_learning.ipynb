```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df = pd.read_csv("student_scores.csv")
# df.head() # commented out to avoid displaying during the fix

# Select the 'Scores' column as the target (y), which can be a Series
target = df['Scores']
# Select the 'Hours' column as the feature (X), ensuring it's a DataFrame
# by passing a list of column names
data = df[['Hours']]

# df.shape # commented out
# data # commented out
# target # commented out

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(data, target, train_size=0.7, random_state=0)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)

# x_train.head() # commented out
# x_test.head() # commented out
# y_train.head() # commented out
# y_test.head() # commented out

# Initialize and train the Linear Regression model
regressor = LinearRegression()
# The fit method will now receive a 2D DataFrame for x_train
regressor.fit(x_train, y_train)
```

```python
print('y intercept:', regressor.intercept_)
print('slope:', regressor.coef_)
```

```python
y_test_pred = regressor.predict(x_test)
```

```python
temp_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred})
temp_df
```

```python
import seaborn as sns
```

```python
sns.histplot(y_test,color = 'blue',alpha =0.5, bins = 5)
sns.histplot(y_test_pred,color = 'red',alpha =0.5, bins = 5)
```

evaluation


```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
```

```python
print(metrics.r2_score(y_test, y_test_pred))
```
