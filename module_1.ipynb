```python
class calc:
  def __init__(self):
    self.value1=0
    self.value2=0
  def get_input(self):
    while True:
      try:
        self.value1=float(input('enter first number'))
        self.value2=float(input('enter second number'))
        return
      except valueError:
        print('please Enter Valid numbers')

  def add(self):
    return self.value1 + self.value2
  def sub(self):
    return self.value1 - self.value2
  def mul(self):
    return self.value1 * self.value2
  def div(self):
    try:
      return self.value1 / self.value2
    except ZeroDivisionError:
        print('division by zero is not possible')
```

```python
# Create an instance of the calc class and assign it to a variable (e.g., calculator)
calculator = calc()

while True:
  print("select an option:")
  print("1.add")
  print("2.sub")
  print("3.mul")
  print("4.div")
  print("5.exit")

  choice = input("enter your choice(1-5):")

  if choice in {'1','2','3','4'}:
    calculator.get_input()  # Call methods on the calculator instance
    if choice == '1':
      result = calculator.add()
      print('result =',result)
    elif choice == '2':
      result = calculator.sub()
      print('result =',result)
    elif choice == '3':
      result = calculator.mul()
      print('result =',result)
    elif choice == '4':
      result = calculator.div()
      print('result =',result)
  else:
    print('invalid choice')

  new_calculation = input("Do you want to perform another calculation? (yes/no): ")
  if new_calculation.lower() != 'yes':
    break
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report
```

```python
data=load_breast_cancer()
x=data.data
y=data.target
```

```python
from sklearn.pipeline import Pipeline # Import the Pipeline class

# ... (your other imports)

pipeline = Pipeline([('scaler', StandardScaler()), ('svm', SVC())]) # Use Pipeline to create a pipeline object
```

```python
Parameters = {'svm__C':[0,1,1,10],'svm__kernel':['linear','rbf']}
```

```python
grid_search = GridSearchCV(pipeline,Parameters,cv=5)
grid_search.fit(x_train,y_train)
```

```python
best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
print(accuracy)
```

```python
classification_rep = classification_report(y_test,y_pred)
print(classification_rep)
```

```python
cross_val_scores = cross_val_score(best_model,x,y,cv=5)
mean_accuracy = cross_val_scores.mean()
print(mean_accuracy)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.datasets import fetch_california_housing
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print(mse)
print(r2)
```

```python
import numpy as np
```

```python
X_train_scaled=np.concatenate((np.ones((X_train_scaled.shape[0],1)),X_train_scaled),axis=1)
X_test_scaled=np.concatenate((np.ones((X_test_scaled.shape[0],1)),X_test_scaled),axis=1)
```

```python
def Linear_regression(x,y):
  weights=np.linalg.inv(x.T @ x)@x.T @ y
  return weights

weights = Linear_regression(X_train_scaled,y_train)
y_pred = X_test_scaled @ weights
```

```python
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print(mse)
print(r2)
```

```python
import gym
import torch
import torch.nn as nn
import torch.optim as opt

```

```python
env=gym.make('CartPole-v1')
```

```python
model = nn.Sequential(
    nn.Linear(env.observation_space.shape[0], 64), # Access the shape attribute using .shape[0]
    nn.ReLU(),
    nn.Linear(64,64),
    nn.ReLU(),
    nn.Linear(64,env.action_space.n)
)
```

```python
optimizer = opt.Adam(model.parameters(),lr=0.001)
criterion = nn.CrossEntropyLoss()
```

```python
!pip install numpy --upgrade
```

```python
n=100
for episode in range(n):
  state=env.reset()
  done=False
  tr=0
  while not done:
    action_probs=torch.softmax(model(torch.tensor(state,dtype=torch.float32)),dim=-1)
    action=torch.multinomial(action_probs,1).item()
    next_state,reward,done,_=env.step(action)
    optimizer.zero_grad()
    loss=criterion(model(torch.tensor(state,dtype=torch.float32)),torch.tensor(action))
    loss.backward()
    optimizer.step()
    state=next_state
    tr+=reward
  print(f"episode:{episode+1},total reward:{tr}")

  env.close()
```
